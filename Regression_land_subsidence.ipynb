```python
import pandas as pd
from pandas import set_option
data = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/ST_FE_CU_SPECIFIC.csv')
print(data.shape)

set_option('display.max_rows', 2490)
set_option('display.max_columns', 32)
data.head(5)
```

```python
data = data.dropna(axis = 0)
data.isnull().sum().sum()
```

```python
data.corr()
```

```python
X = data.drop('D20140322',axis = 1)
Y = data['D20140322']
```

```python
from sklearn.model_selection import train_test_split

# Train and test split
test_size = 0.2
seed = 42
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,
                                                    test_size = test_size,
                                                    random_state = seed)
print("X_train shape:", X_train.shape)
print("Y_train shape:", Y_train.shape)
print("X_test shape:", X_test.shape)
print("Y_test shape:", Y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)
from sklearn import metrics
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, y_pred))
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, y_pred))
print("R_squared: ", metrics.r2_score(Y_test, y_pred))
```

# **Tuning ExtraTreesRegressor**

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import ExtraTreesRegressor
X = data.drop('D20140322',axis = 1)
Y = data['D20140322']
scaler = StandardScaler().fit(X)
rescaledX = scaler.transform(X)

test_size = 0.2
seed = 42
X_train, X_test, Y_train, Y_test = train_test_split(rescaledX, Y, test_size=test_size, random_state=seed)

EX_clf = ExtraTreesRegressor()
parameters = {

    'max_depth': [None , 3 , 5 , 7 ,10 , 15],
    'min_samples_split': [2 , 3, 4 , 6 , 7 , 8 ],
    'n_estimators': [50, 80, 100, 150, 200 , 220 , 300],
    'bootstrap' : [True , False]

}
grid = GridSearchCV(EX_clf, param_grid=parameters, cv=5, scoring='r2')
model_grid = grid.fit(X_train, Y_train)

print('Best hyperparameters are: '+str(model_grid.best_params_))
print('Best score is: '+str(model_grid.best_score_))
```

**Tuning usually fix the trade-off between overfiting and underfiting  .  Lets check:**

```python
model = ExtraTreesRegressor(bootstrap= True, max_depth= 15, min_samples_split= 2, n_estimators= 220)
model.fit(X_train, Y_train)
y_pred = model.predict(X_train)


from sklearn import metrics
print('Mean Squared Error:', metrics.mean_squared_error(Y_train, y_pred))
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_train, y_pred))
print("R_squared: ", metrics.r2_score(Y_train, y_pred))
```

```python
model = ExtraTreesRegressor(bootstrap= True, max_depth= 15, min_samples_split= 2, n_estimators= 220 )
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)


from sklearn import metrics
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, y_pred))
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, y_pred))
print("R_squared: ", metrics.r2_score(Y_test, y_pred))
```
