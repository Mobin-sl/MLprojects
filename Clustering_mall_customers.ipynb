```python
stingTOnumeric = {
    'Male':1,
    'Female':0
}
data["Gender"] =  data["Gender"].map(stingTOnumeric)
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
f1 = data['Gender'].values
f2 = data['Age'].values
f3 = data['Annual Income (k$)'].values
f4 = data['Spending Score (1-100)'].values

X = np.array(list(zip(f1, f2 , f3 , f4)))
```

```python
from sklearn.cluster import KMeans
K = range(1, 15)
sum_of_squared_distances = []

for k in K:
  kmeans = KMeans(n_clusters = k , n_init=10)
  kmeans = kmeans.fit(X)
  sum_of_squared_distances.append(kmeans.inertia_)
```

```python
plt.plot(K, sum_of_squared_distances, 'bx-')
```

```python
from sklearn.cluster import KMeans

# Initialize the class object
kmeans = KMeans(n_clusters = 4 , n_init=10)

# Predict the labels of clusters
label = kmeans.fit_predict(X)
label
```

```python
ziro = list(label).count(0)
one = list(label).count(1)
two = list(label).count(2)
three = list(label).count(3)
```

```python
x_axis = []
x_axis.append(ziro)
x_axis.append(one )
x_axis.append(two )
x_axis.append(three)

colors = ['pink','yellow' ,'darkorange' ,'red']

print("No.of.diabetic patients : " + str(x_axis[0]) +
      "\nNo.of.non-diabetic patients : " + str(x_axis[1]))

y = ['ziro' , 'one' , 'two' , 'three']

plt.pie(x_axis, labels=y, colors=colors, shadow='true')
plt.show()
```
